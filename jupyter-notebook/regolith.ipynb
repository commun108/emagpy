# Interpolation options
In this notebook you will lear how to use the python API of emagpy.

First let's start with an example where spatial EMI data were collected on a thin soil (30 cm thick) and below the regolith which contains mostly clay.

```python
import os
import sys
sys.path.insert(0,'../src/')
testdir = '../src/examples/saprolite/'
from emagpy import Problem
```

```python
k = Problem() # create an emagpy instance
k.createSurvey(testdir + 'regolith.csv')
```

```python
k.show()
```

```python
k.convertFromNMEA() # convert NMEA string to British Grid (by default)
k.showMap()
```

```python
k.showMap(contour=True, pts=True) # do the contouring
```

Now one can use the different interpolation method to get a map. Let's compare them below.

```python
k = Problem() # create an emagpy instance
k.createSurvey(testdir + 'regolith.csv')
k.convertFromNMEA()
k.gridData(method='nearest')
k.surveys[0].df = k.surveys[0].dfg
k.showMap()
```

```python
k = Problem() # create an emagpy instance
k.createSurvey(testdir + 'regolith.csv')
k.convertFromNMEA()
k.gridData(method='linear')
k.surveys[0].df = k.surveys[0].dfg
k.showMap()
```

```python
k = Problem() # create an emagpy instance
k.createSurvey(testdir + 'regolith.csv')
k.convertFromNMEA()
k.gridData(method='cubic')
k.surveys[0].df = k.surveys[0].dfg
k.showMap()
```

```python
k = Problem() # create an emagpy instance
k.createSurvey(testdir + 'regolith.csv')
k.convertFromNMEA()
k.gridData(method='idw')
k.surveys[0].df = k.surveys[0].dfg
k.showMap()
```

## Observations
Cubic interpolation seems to perform the best and IDW has this problem of small mountains around the points. This might be solved by changing the exponent of the inverpolation in the function of IDW. Note that we haven't compare it to kriging yet.
